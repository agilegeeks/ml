```python
from pprint import pprint
import drest
api_url = 'http://orchestrator.rotld.ro:9090/api/v1'
```

```python
req = drest.request.TastyPieRequestHandler()
req.add_param('query','increase(mdw_whois_hits{service="whois_pub"}[15m])')
req.add_param('start','2017-03-03T10:10:00.000Z')
req.add_param('end','2017-03-04T10:10:00.000Z')
req.add_param('step','15m')
```

```python
response = req.make_request('GET', api_url+'/query_range/')
```

```python
results = response.data['data']['result']
pprint(results)
vfound = 0
vnotfound = 0 
for entry in results:
    status = entry['metric']['status']
    if status=='not_found':
        vnotfound = float(entry['value'][1])
    elif status=='found':
        vfound = float(entry['value'][1])

#vfound = vfound/1000
#vnotfound = vnotfound/1000
print ("\n\n\n\nFound:%.8f   Not Found:%.8f " % (vfound,vnotfound))
```
